#Working with Databases in Python and Pandas
<font color = red> **Important!** </font> First, download SQLlite database called [flights.db](https://drive.google.com/file/d/1T2Nf5vGq6fJfM7x_U8N5PwZVtzZvteru/view?usp=sharing).
```
import pandas as pd
import sqlite3
conn = sqlite3.connect("flights.db")
conn
```

The database contains three tables â€” `airports`, `airlines`, and `routes`. In Python Pandas, you can do a SQL Query and save the results in a data frame like below!

```
airlines = pd.read_sql_query("SELECT * FROM airlines;", conn)
airlines
```

```
airports = pd.read_sql_query("SELECT * FROM airports;", conn)
airports
```

<font color=red>Write code that reads in the routes table into a variable called `routes` using a SELECT * query.</font>

```
routes = pd.read_sql_query("SELECT * FROM routes;", conn)
routes
```

Let's try a SQL query with a WHERE clause to pull out just the airports in the United States.

```
us_airports = pd.read_sql_query("SELECT * FROM airports WHERE country = 'United States';", conn)
us_airports
```

<font color=red>Write a SQL query like above that prints out the **airports** for the city Boston saved in a variable for `boston_airports`.</font>

```
boston_airports = pd.read_sql_query("SELECT * FROM airports WHERE city = 'Boston';", conn)
boston_airports
```

<font color=red>Write a SQL query like above that prints out the **routes** where the source is 'BOS' (all flights from Boston).</font>

```
boston_routes = pd.read_sql_query("SELECT * FROM routes WHERE source = 'BOS';", conn)
boston_routes
```

You may have noticed that the `routes` table has foreign keys for `airline_id` and `source_id` and `dest_id` instead of the airlines and airport names. You have to look these up in the other tables. Here's a SQL query that joins together the routes and the airports and the airlines tables so that you can see which airlines fly to which airports.

```
flights = pd.read_sql_query("SELECT * FROM routes JOIN airports ON airports.id = routes.dest_id JOIN airlines ON airlines.id = routes.airline_id", conn)
flights
```
