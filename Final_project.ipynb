```python
import pandas as pd
#link to kaggle dataset: https://www.kaggle.com/datasets/abecklas/fifa-world-cup/
```

1- Importing the content of the dataset World Cups

```python
fifa_world_cups_path = 'WorldCups_data/WorldCups.csv'

fifa_world_cups = pd.read_csv(fifa_world_cups_path)

fifa_world_cups['Country'].replace('Germany FR', 'Germany', inplace=True)

fifa_world_cups['Winner'].replace('Germany FR', 'Germany', inplace=True)

fifa_world_cups.head()
```

2- Importing all the World cup matches

```python
fifa_world_cups_matches_path = 'WorldCups_data/WorldCupMatches.csv'

fifa_world_cups_matches = pd.read_csv(fifa_world_cups_matches_path)

fifa_world_cups_matches['Home Team Name'].replace('Germany FR', 'Germany', inplace=True)

fifa_world_cups_matches['Away Team Name'].replace('Germany FR', 'Germany', inplace=True)

fifa_world_cups_matches.head()
```

3- Importing all the World cup players




```python
fifa_world_cups_players_path = 'WorldCups_data/WorldCupPlayers.csv'

fifa_world_cups_players = pd.read_csv(fifa_world_cups_players_path)

fifa_world_cups_players.head()
```

**Question 1:** Does the host country have a higher percentage chance to win

---

the world cup?

```python
Number_of_Worldcups = fifa_world_cups.Country.count()

winners = fifa_world_cups.Winner
winners

host = fifa_world_cups.Country
host
i=0
host_with_a_cup = 0
for i in range(Number_of_Worldcups):
  if(host[i] == winners[i]):
    host_with_a_cup += 1

percentage = (host_with_a_cup / Number_of_Worldcups)

percentage
```

**Question 2**: Does the team  that scores more goals always win the world cup?


```python
world_cup_years = fifa_world_cups_matches.Year.unique()
world_cup_years[:Number_of_Worldcups]
# winner_with_most_score is a variable that will indicate the amount of winners who had the most score
winner_with_most_score = 0
for year in world_cup_years[:Number_of_Worldcups]:
  world_cup_cur = fifa_world_cups_matches[fifa_world_cups_matches.Year == year]
  sum_name = world_cup_cur.groupby("Home Team Name")[["Home Team Goals","Away Team Goals"]].apply(sum)
  total = sum_name.sum(axis=1)
  if(len(total.index[total==total.max()])==1):
    if(total.index[total==total.max()] == fifa_world_cups.Winner[fifa_world_cups.Year==year]):
      winner_with_most_score += 1
  else:
    size = len(total.index[total==total.max()])
    win = fifa_world_cups.Winner[fifa_world_cups.Year==year]
    if(win.values in total.index[total==total.max()].values):
        winner_with_most_score += 1
print("Times were the winner had the most score " + str(winner_with_most_score))
winner_with_most_score/Number_of_Worldcups
```

**Question 3**: If a country has already won one world cup what's the percentage chance to win another one?


```python
winners = fifa_world_cups.groupby('Winner').count()
winners_with_more_than_one_cup = winners[winners.Country >1]
# winners_with_more_than_one_cup
winners_with_more_than_one_cup.Year.sum()
percentage = (winners_with_more_than_one_cup.Year.sum() / Number_of_Worldcups)
percentage
```
